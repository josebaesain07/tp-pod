Combine (*join*) las tablas `airlines`, `airports`, y `flights` en una sola tabla.

```python
#Escribir el codigo para responder la pregunta en esta celda
import pandas as pd
airlines = pd.read_csv('/content/airlines.csv')
airports = pd.read_csv('/content/airports.csv')
flights = pd.read_csv('/content/flights.csv')
merged_df = pd.merge(flights, airlines, left_on='AIRLINE', right_on='IATA_CODE', how='left')
final_df = pd.merge(merged_df, airports, left_on='ORIGIN_AIRPORT', right_on='IATA_CODE', how='left')
final_df = final_df.rename(columns={'AIRLINE_x': 'AIRLINE', 'IATA_CODE_x': 'AIRLINE_IATA', 'IATA_CODE_y': 'AIRPORT_IATA'})
final_df
```

¿Cuantos vuelos fueron al aeropuerto JFK?

```python
#Escribir el codigo para responder la pregunta en esta celda
(final_df['DESTINATION_AIRPORT'] == 'JFK').sum()
```

¿Cuantos vuelos hizo la aerolinea AA?

```python
#Escribir el codigo para responder la pregunta en esta celda
vuelos_AA = final_df[final_df['AIRLINE'] == 'AA']
print(len(vuelos_AA))
```

¿Que aerolineas (las primeras 5) tuvo la menor cantidad de vuelos con atrasos? Imprimirlas.

```python
#Escribir el codigo para responder la pregunta en esta celda
vuelos_atrasados = final_df[final_df['DEPARTURE_DELAY'] > 0]
vuelos_atrasados_agrupados = vuelos_atrasados.groupby('AIRLINE')
vuelos_atrasados_agrupados_tamanio = vuelos_atrasados_agrupados.size()
vuelos_atrasados_agrupados_tamanio = vuelos_atrasados_agrupados_tamanio.sort_values(ascending=True)

print(vuelos_atrasados_agrupados_tamanio.head())
```

¿Que aerolineas (las primeras 5) tuvo la mayor cantidad de vuelos con atrasos? Imprimirlas.

```python
#Escribir el codigo para responder la pregunta en esta celda
vuelos_atrasados_agrupados_tamanio = vuelos_atrasados_agrupados_tamanio.sort_values(ascending=False)
print(vuelos_atrasados_agrupados_tamanio.head())
```

Haga un resumen de las razones por la cual los vuelos se atrasan.

```python
#Escribir el codigo para responder la pregunta en esta celda
delay_reasons = flights[['AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY',
                        'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY']].sum()

print("Summary of Delay Reasons:")
print(delay_reasons)

total_delays = delay_reasons.sum()
delay_percentages = (delay_reasons / total_delays) * 100
print("\nDelay Reasons with Percentages:")
print(delay_percentages)
```
